# **Task 1: Conceptual Understanding**

# **1. What is the difference between "Love" and "love" in NLP?**

Answer :-  In NLP, most systems are case-sensitive by default. This means "Love" (with a capital 'L') and "love" (all lowercase) are treated as two entirely different tokens or unique words.


*   "Love": Might be recognized as a proper noun (like a name) or the start of a sentence.
*   "love": Generally recognized as a common noun or a verb.

# **2. What happens if stopwords are not removed?**

Answer :- Stopwords are common words like "the," "is," "and," or "in" that carry little unique meaning. If they are not removed:


*   *Increased Noise* : The model may focus on these frequent but irrelevant words instead of the "keywords" that actually define the topic.
*   *Computational Cost* : The dataset remains much larger, meaning the computer needs more memory and more time to process the information.
*   *Skewed Accuracy* : In tasks like document classification, the high frequency of these words can confuse the model, leading to less accurate results.



# **3. Provide two real-world scenarios where removing stopwords can be harmful.**

Answer :- While removing stopwords is common, it can be a mistake in specific contexts:
*   *Sentiment Analysis* : Small words can change the entire meaning of a sentence. For example, in the sentence "I do not like this," removing the stopword "not" changes the sentiment from negative to positive.
*  *Language Translation* : Stopwords are essential for the grammatical structure and "flow" of a language. If you remove them before translating, the output would likely be a "word salad" that doesn't make sense to a native speaker.

# **4. Explain the difference between stemming and lemmatization.**

Answer :-
*  Stemming and lemmatization are NLP text normalization techniques that reduce words to a base form.
*  Stemming crudely chops off suffixes (e.g., "studies" -> "studi") fast, often producing non-words.
*  Lemmatization uses vocabulary/morphological analysis to return a valid dictionary word (e.g., "studies" -> "study"), ensuring semantic accuracy.

# **Task 2: Build Advanced Preprocessing Function**

```python
import re

def preprocess_text(text):
    if not isinstance(text, str):
        return ""

    # 4. Convert text to lowercase
    text = text.lower()

    # 6. Remove URLs and email-like patterns
    url_pattern = r'https?://\S+|www\.\S+'
    email_pattern = r'\S+@\S+\.\S+'
    text = re.sub(url_pattern, '', text)
    text = re.sub(email_pattern, '', text)

    # Implicit from Examples 3 & 4 (WOW!!! -> wow): Remove punctuation
    text = re.sub(r'[^\w\s]', ' ', text)

    # 1. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 3. Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # 5. Remove very short tokens & 2. Remove extra spaces
    exceptions = {'no', 'not'}
    words = text.split()

    filtered_words =[]
    for word in words:
        # Keep if length is > 2 OR if the word is in our exceptions list
        if len(word) > 2 or word in exceptions:
            filtered_words.append(word)

    # Rejoin the processed words into a single string with single spaces
    final_text = ' '.join(filtered_words)

    return final_text

# TEST CASES BASED ON EXAMPLES
if __name__ == "__main__":
    test_cases =[
        "I have 2 dogs",
        "This is   good",
        "soooo goooood!!!",
        "WOW!!! This is GREAT!!!",
        "Do not say no to this",
        "Visit http://example.com or test@email.com now"
    ]

    for test in test_cases:
        print(f"Original: '{test}'")
        print(f"Processed: '{preprocess_text(test)}'\n")
```

# **Task 3 till Task 7 - Stress Testing, Token Analytics, Frequency Analysis,  Build Full Pipeline, Error Handling**

```python
import re
from collections import Counter


# TASK 2 & TASK 7: ADVANCED PREPROCESSING & ERROR HANDLING
def preprocess_text(text):

    # TASK 7: Error Handling - Type checking and Empty string handling
    if not isinstance(text, str):
        return ""
    if not text.strip():
        return ""

    try:
        # TASK 2: 4. Convert text to lowercase
        text = text.lower()

        # TASK 2: 6. Remove URLs and email-like patterns
        text = re.sub(r'https?://\S+|www\.\S+', '', text)
        text = re.sub(r'\S+@\S+\.\S+', '', text)

        # Remove punctuation (and TASK 7: Emojis) by keeping only alphanumeric and whitespace
        text = re.sub(r'[^\w\s]', ' ', text)

        # TASK 2: 1. Remove numbers (and TASK 7: Handle numbers-only inputs)
        text = re.sub(r'\d+', '', text)

        # TASK 2: 3. Handle repeated characters (reduce 3+ to 2)
        text = re.sub(r'(.)\1{2,}', r'\1\1', text)

        # TASK 2: 5. Remove short tokens & 2. Remove extra spaces
        exceptions = {'no', 'not'}
        words = text.split()

        filtered_words =[word for word in words if len(word) > 2 or word in exceptions]

        return ' '.join(filtered_words)

    except Exception as e:
        return ""



# TASK 6: BUILD FULL PIPELINE
def full_pipeline(text_list):
    all_combined_tokens =[]
    clean_sentences =[]

    for text in text_list:
        cleaned_text = preprocess_text(text)
        tokens = cleaned_text.split() if cleaned_text else[]

        all_combined_tokens.extend(tokens)
        clean_sentences.append(cleaned_text)

    return {
        "tokens": all_combined_tokens,
        "clean_sentences": clean_sentences
    }


# MAIN EXECUTION
if __name__ == "__main__":

    # Sample inputs provided in the assignment
    sample_inputs =[
        "Get 100% FREE access now!!!",
        "I absolutely looooved this product 😍😍",
        "Worst service ever... 0/10",
        "Call me at 9876543210",
        "This is THE best course!!!",
        "Visit https://openai.com now!",
        "Nooooo this is baaad!!!",
        "OK OK OK I got it",
        "Win $$$ now!!! Limited offer!!!",
        "I am not happy with this"
    ]

    print("\n" + "="*50)
    print("TASK 7: ERROR HANDLING DEMONSTRATION")
    print("="*50)
    edge_cases =["", "😍😍🚀", "12345 67890", None, 99.9]
    for case in edge_cases:
        print(f"Input: {repr(case):<15} | Cleaned: repr({repr(preprocess_text(case))})")


    print("\n" + "="*50)
    print("TASK 3: STRESS TESTING")
    print("="*50)
    for sentence in sample_inputs:
        cleaned_sentence = preprocess_text(sentence)
        cleaned_tokens = cleaned_sentence.split() if cleaned_sentence else[]

        print(f"Original Text:    {sentence}")
        print(f"Cleaned Tokens:   {cleaned_tokens}")
        print(f"Cleaned Sentence: {cleaned_sentence}")
        print("-" * 50)


    print("\n" + "="*50)
    print("TASK 4: TOKEN ANALYTICS")
    print("="*50)
    print(f"{'Original Sentence':<43} | {'Total':<5} | {'Unique':<6} | {'Avg Len':<7}")
    print("-" * 70)

    for sentence in sample_inputs:
        cleaned_tokens = preprocess_text(sentence).split()

        # 1. Total tokens, 2. Unique tokens, 3. Average token length
        total_tokens = len(cleaned_tokens)
        unique_tokens = len(set(cleaned_tokens))
        avg_length = sum(len(t) for t in cleaned_tokens) / total_tokens if total_tokens > 0 else 0.0

        print(f"{sentence[:40]:<43} | {total_tokens:<5} | {unique_tokens:<6} | {avg_length:.2f}")


    print("\n" + "="*50)
    print("TASK 6: RUNNING FULL PIPELINE")
    print("="*50)
    # Execute the pipeline to get the dictionary containing all tokens
    pipeline_results = full_pipeline(sample_inputs)
    print(f"Pipeline successfully returned dictionary with keys: {list(pipeline_results.keys())}")
    print(f"Total Combined Tokens extracted: {len(pipeline_results['tokens'])}")
    print(f"Total Clean Sentences processed: {len(pipeline_results['clean_sentences'])}")


    print("\n" + "="*50)
    print("TASK 5: FREQUENCY ANALYSIS")
    print("="*50)
    # Use the token list generated by Task 6's pipeline
    all_tokens = pipeline_results["tokens"]
    token_counts = Counter(all_tokens)

    top_10 = token_counts.most_common(10)
    bottom_5 = token_counts.most_common()[-5:]

    print("Top 10 Most Frequent Words:")
    for i, (word, count) in enumerate(top_10, 1):
        print(f"  {i}. '{word}' ({count})")

    print("\nTop 5 Least Frequent Words:")
    for i, (word, count) in enumerate(bottom_5, 1):
        print(f"  {i}. '{word}' ({count})")
    print("="*50 + "\n")
```
