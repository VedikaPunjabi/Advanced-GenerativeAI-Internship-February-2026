```
# ==============================================================================
# ASSIGNMENT NLP – 3 (Chatbot using Transformers)
# Description: A console-based chatbot using Hugging Face Transformers.
# Model Used: facebook/blenderbot-400M-distill [Using BlenderBot as it provides a great balance of conversational chit-chat and factual answering (unlike standard GPT-2).]
# ==============================================================================

# Import required libraries
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import transformers
import warnings

# Suppress Hugging Face info/warnings for a completely clean console output
warnings.filterwarnings('ignore')
transformers.logging.set_verbosity_error()

def run_chatbot():
    """
    Initializes and runs an interactive console-based chatbot.
    Matches the Expected Pipeline Flow:
    User Input -> Model Processing -> Response Generation -> Display Output -> Loop
    """

    # 1. Model Loading
    model_name = "facebook/blenderbot-400M-distill"
    print("Loading AI model... Please wait.\n")

    # Load tokenizer and pre-trained transformer model
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

    # Initialize a list to maintain conversation flow/history
    conversation_history = []

    # Expected startup message
    print("Chatbot: Hello! I am your AI assistant. How can I help you today?")

    # 4. Continuous Conversation Loop
    while True:
        # 2. User Input Handling
        user_input = input("User: ")

        # 5. Exit Condition
        # Stop the chatbot cleanly if user types 'exit' or 'quit'
        if user_input.strip().lower() in ['exit', 'quit']:
            print("Chatbot ends the conversation.")
            break

        # Ignore empty inputs to prevent the model from crashing
        if not user_input.strip():
            continue

        # Append the new user input to the conversation history
        conversation_history.append(user_input)

        # Keep only the last 3 exchanges to prevent memory overload
        # and keep the chatbot focused on the current topic
        context_string = "  ".join(conversation_history[-3:])

        # 3. Response Generation
        # Tokenize the input context and convert to PyTorch tensors
        inputs = tokenizer(context_string, return_tensors="pt")

        # Generate a response using the transformer model
        response_ids = model.generate(
            **inputs,
            max_length=150,
            pad_token_id=tokenizer.eos_token_id,
            temperature=0.7,   # Controls randomness (lower = more focused)
            do_sample=True     # Allows dynamic generation
        )

        # Decode the model's response back into human-readable text
        bot_response = tokenizer.decode(response_ids[0], skip_special_tokens=True)

        # Add the bot's response to the history so it remembers what it said
        conversation_history.append(bot_response)

        # Display Output
        print(f"Chatbot: {bot_response}")

# Execute the chatbot
if __name__ == "__main__":
    run_chatbot()
```
