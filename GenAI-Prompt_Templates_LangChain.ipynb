**Task 1: Converting Hardcoded Prompts**

```
#Import Langchain
from langchain_core.prompts import PromptTemplate

# Task 1
template_text = "Explain {topic} in simple terms for beginners"

prompt_template = PromptTemplate(
    input_variables=["topic"],
    template=template_text
)

print(prompt_template.format(topic="Machine Learning"))
```

**Task 2: Multi-Input Prompt System**

```
# Template defined in documentation
multi_input_template = "Explain {topic} for {audience} in a {tone} tone"

prompt_multi = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=multi_input_template
)

# Test Case from source
print(prompt_multi.format(topic="Python", audience="kids", tone="fun"))
```

**Task 4: ChatPrompt Template System**

```
from langchain_core.prompts import ChatPromptTemplate

# Initialize the chat-based system
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful {role}."), # Behavior
    ("human", "Explain {topic} in detail.")  # User input
])

# Testing for role: teacher and topic: Neural Networks
print(chat_template.format_messages(role="teacher", topic="Neural Networks"))
```

**Task 5: Input Validation Layer**

```
def validate_inputs(audience, tone):
    # Defining allowed values per source
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # Assigning default values if invalid
    final_audience = audience if audience in valid_audiences else "beginner"
    final_tone = tone if tone in valid_tones else "formal"

    return final_audience, final_tone
```

**Task 6: Prompt Generator App**

```
def generate_prompt(topic, audience, tone, style):
    # First, validate the inputs
    v_audience, v_tone = validate_inputs(audience, tone)

    # Select template based on style
    if style == "teaching":
        template = "Explain {topic} clearly step by step for {audience} in a {tone} tone"
    elif style == "interview":
        template = "Ask 3 interview questions about {topic} for an {audience} in a {tone} tone"
    else: # Storytelling
        template = "Explain {topic} as a story for {audience} in a {tone} tone"

    # Use PromptTemplate for generation
    prompt_obj = PromptTemplate(input_variables=["topic", "audience", "tone"], template=template)
    return prompt_obj.format(topic=topic, audience=v_audience, tone=v_tone)

# Example output
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

**Task 7: Template Reusability Test**

```
# Task 7: Run ONE template with 5 different inputs
topics = ["Cloud Computing", "Blockchain", "Data Science", "Cybersecurity", "IoT"]

for t in topics:
    # Using the prompt_multi template created in Task 2
    print(prompt_multi.format(topic=t, audience="expert", tone="formal"))
```
